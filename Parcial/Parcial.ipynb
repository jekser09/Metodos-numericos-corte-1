```python
import numpy as np

productos = np.array([["Sucursal","Manos libres","Celulares","Portatiles","Relojes","Accesorios"],
                      ["Calles26",2,3,20,12,5]])

print("productos:\n")
for i in range(len(productos[0])):
  print(productos[0,i],": ",productos[1,i])
print("\n")

bandera1=False
while bandera1==False:

  p=str.capitalize(input("Ingrese el nombre del producto que desea consultar\n"))  

  for i in range(len(productos[0])):
    if productos[0,i]==p:
      print(productos[0,i],":",productos[1,i])
      if int(productos[1,i])<=5:
        print("La cantidad de productos es muy baja por favor indique cuantos desea añadir")

        bandera2=False
        while bandera2==False:
          c1=int(input())
          if c1<=0:
            print("ingrese un valor valido")
            bandera2=False
          else:
            c1=c1+int(productos[1,i])
            bandera2=True  
        productos[1,i]=c1
        print("Cantidad de ",productos[0,i],": ",productos[1,i])
        print("tabla actualizada:\n")
        for i in range(len(productos[0])):
          print(productos[0,i],": ",productos[1,i])
        print("\n")
        bandera1=True
        break
      bandera1=True  
      break
   
    


```
